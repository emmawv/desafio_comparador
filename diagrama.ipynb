```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.sankey import Sankey

# Crear la figura
fig, ax = plt.subplots(figsize=(10, 8))

# Dibujar los rectángulos y las flechas
ax.text(0.5, 0.95, 'Inicio', ha='center', va='center', fontsize=12, bbox=dict(facecolor='lightblue', edgecolor='black'))

# Primer bloque (input de datos)
ax.text(0.5, 0.8, '1. Input de datos', ha='center', va='center', fontsize=12, bbox=dict(facecolor='lightgreen', edgecolor='black'))
ax.text(0.2, 0.7, 'Formulario', ha='center', va='center', fontsize=10, bbox=dict(facecolor='lightyellow', edgecolor='black'))
ax.text(0.5, 0.7, 'Subiendo PDF', ha='center', va='center', fontsize=10, bbox=dict(facecolor='lightyellow', edgecolor='black'))
ax.text(0.8, 0.7, 'Foto', ha='center', va='center', fontsize=10, bbox=dict(facecolor='lightyellow', edgecolor='black'))

# Flechas del input de datos
ax.arrow(0.5, 0.9, 0, -0.05, head_width=0.03, head_length=0.02, fc='black', ec='black')
ax.arrow(0.5, 0.8, -0.3, -0.05, head_width=0.03, head_length=0.02, fc='black', ec='black')
ax.arrow(0.5, 0.8, 0, -0.05, head_width=0.03, head_length=0.02, fc='black', ec='black')
ax.arrow(0.5, 0.8, 0.3, -0.05, head_width=0.03, head_length=0.02, fc='black', ec='black')

# Segundo bloque (análisis)
ax.text(0.5, 0.5, '2. Análisis y comparación', ha='center', va='center', fontsize=12, bbox=dict(facecolor='lightgreen', edgecolor='black'))
ax.text(0.5, 0.4, 'Extracción de términos clave', ha='center', va='center', fontsize=10, bbox=dict(facecolor='lightyellow', edgecolor='black'))
ax.text(0.5, 0.35, 'Comparación con tarifas en base de datos', ha='center', va='center', fontsize=10, bbox=dict(facecolor='lightyellow', edgecolor='black'))

# Flecha hacia el análisis
ax.arrow(0.5, 0.65, 0, -0.1, head_width=0.03, head_length=0.02, fc='black', ec='black')

# Tercer bloque (resultados)
ax.text(0.5, 0.2, '3. Resultados del mercado', ha='center', va='center', fontsize=12, bbox=dict(facecolor='lightgreen', edgecolor='black'))
ax.text(0.5, 0.1, 'Mostrar comparativa de precios', ha='center', va='center', fontsize=10, bbox=dict(facecolor='lightyellow', edgecolor='black'))

# Flecha hacia los resultados
ax.arrow(0.5, 0.3, 0, -0.05, head_width=0.03, head_length=0.02, fc='black', ec='black')

# Ajustes finales
ax.axis('off')  # Ocultar ejes

# Mostrar el diagrama
plt.show()
```
