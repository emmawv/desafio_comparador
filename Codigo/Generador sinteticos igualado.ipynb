```python
#CREAMOS LOS REGISTROS SINTETICOS

# diccionario con 104 registros de tarifa con 20 registros cada una de total de kWh consumido en el periodo de facturacion
# Mantiene semilla y nombre del doc si no se incluye en la declaración de la función


def generador_sint(num_reg_ppal, num_reg_sec, file_name='../Datos/consumos_modelo.csv', seed=42):
    # Establecer la semilla
    np.random.seed(seed)

    # Generar el diccionario de consumos
    consumos = {}

    for tarifa in range(num_reg_ppal):
        consumos[tarifa] = []
        for registro in range(num_reg_sec):
            valores = np.random.randint(100, 300)
            consumos[tarifa].append(valores)

    # Generar pesos para 3 tramos
    pesos_3t = []
    for tarifa in range(num_reg_ppal):
        for _ in range(num_reg_sec):
            a, b = np.random.randint(0, 101, 2)
            c = 100 - a - b
            while c < 0:
                a, b = np.random.randint(0, 101, 2)
                c = 100 - a - b
            pesos_3t.append([a, b, c])

    # Generar pesos para 2 tramos
    pesos_2t = []
    for tarifa in range(num_reg_ppal):
        for _ in range(num_reg_sec):
            d = np.random.randint(0, 101)
            e = 100 - d
            pesos_2t.append([d, e])

    # Aplanar el diccionario de consumos en una lista
    registros = []
    for tarifa, valores in consumos.items():
        for valor in valores:
            registros.append({'tarifa': tarifa, 'consumo_kWh': valor})

    # Crear el DataFrame
    totales = pd.DataFrame(registros)

    # Añadir columnas de tramos con los pesos generados
    totales['%consumo_valle'] = [p[0] for p in pesos_3t]
    totales['%consumo_llano'] = [p[1] for p in pesos_3t]
    totales['%consumo_punta'] = [p[2] for p in pesos_3t]
    totales['%consumo_8h'] = [p[0] for p in pesos_2t]
    totales['%consumo_16h'] = [p[1] for p in pesos_2t]

    # Guardar el DataFrame en el archivo CSV
    totales.to_csv(file_name, index=False)
```

```python
# declaramos como variable la nueva base de datos mergeada

dt = pd.read_csv('../Datos/copy_clientes_tarifas_modelo.csv')
```

```python
#comprobamos cuantos si hay nan
dt[dt['margen_euros/kWh/dia'].isna()]
```

```python
#comprobamos si no hay nan
dt[dt['margen_euros/kWh/dia'].notna()]
```

```python
# Sustituir los nan por 0
dt['margen_euros/kWh/dia'] = dt['margen_euros/kWh/dia'].fillna(0)
```

```python
# para calcular los consumos reales basandonos en los porcentajes de uso

dt['consumo_unico'] = dt.apply(lambda row: (row['consumo_kWh']) * 1 if row['consumo_kWh'] > 0 else 0, axis=1)
dt['consumo_valle'] = dt.apply(lambda row: (row['consumo_kWh']) * 0.33 if row['consumo_kWh'] > 0 else 0, axis=1)
dt['consumo_llano'] = dt.apply(lambda row: (row['consumo_kWh']) * 0.33 if row['consumo_kWh'] > 0 else 0, axis=1)
dt['consumo_punta'] = dt.apply(lambda row: (row['consumo_kWh']) * 0.33 if row['consumo_kWh'] > 0 else 0, axis=1)
dt['consumo_8h'] = dt.apply(lambda row: (row['consumo_kWh']) * 0.5 if row['consumo_kWh'] > 0 else 0, axis=1)
dt['consumo_16h'] = dt.apply(lambda row: (row['consumo_kWh']) * 0.5 if row['consumo_kWh'] > 0 else 0, axis=1)
```

```python
# para calcular el precio de los consumos basandonos en los porcentajes de uso
dt['consumo_euros_unico'] = dt.apply(lambda row: (row['consumo_kWh'] ) * 1 if row['termino_consumo_euros/kwh/hora'] > 0 else 0, axis=1)
dt['consumo_euros_valle'] = dt.apply(lambda row: (row['consumo_kWh'] * 0.33) * row['tc_valle_euros/kwh/hora'] if row['tc_valle_euros/kwh/hora'] > 0 else 0, axis=1)
dt['consumo_euros_llano'] = dt.apply(lambda row: (row['consumo_kWh'] * 0.33) * row['tc_llano_euros/kwh/hora'] if row['tc_llano_euros/kwh/hora'] > 0 else 0, axis=1)
dt['consumo_euros_punta'] = dt.apply(lambda row: (row['consumo_kWh'] * 0.33) * row['tc_punta_euros/kwh/hora'] if row['tc_punta_euros/kwh/hora'] > 0 else 0, axis=1)
dt['consumo_euros_8h'] = dt.apply(lambda row: (row['consumo_kWh'] * 0.5) * row['tc_8h_euros/kwh/hora'] if row['tc_8h_euros/kwh/hora'] > 0 else 0, axis=1)
dt['consumo_euros_16h'] = dt.apply(lambda row: (row['consumo_kWh'] * 0.5) * row['tc_16h_euros/kwh/hora'] if row['tc_16h_euros/kwh/hora'] > 0 else 0, axis=1)
```

```python
# Calcular el precio final

dt['precio_final'] = dt['consumo_euros_unico'] + dt['consumo_euros_valle'] + dt['consumo_euros_llano'] + dt['consumo_euros_punta'] + dt['consumo_euros_8h'] + dt['consumo_euros_16h']
```
